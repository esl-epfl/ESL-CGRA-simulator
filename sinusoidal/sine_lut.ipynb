# Sine LUT on ESL-CGRA

```python
import sys, math
sys.path.append("../src/")
from cgra import *

KERNEL = "sine_approx"
SCALE = 10000
OUTPUT_ADDR = 10000

load_addrs = [0, 4, 0, 0]
store_addrs = [10000, 10004, 10008, 10012]

run(
    KERNEL,
    pr=["ROUT", "OPS", "R0", "R1", "ALL_LAT_INFO", "ALL_PWR_EN_INFO"],
    load_addrs=load_addrs,
    store_addrs=store_addrs,
    limit=500,
)
```

## Result

```python
with open(os.path.join(KERNEL, "memory_out.csv")) as f:
    mem = {int(r[0]): int(r[1]) for r in csv.reader(f)}

result = mem[OUTPUT_ADDR]
x_test = 4.3
print(f'Result: {result} (={result/SCALE:.6f})')
print(f'Expect: {round(math.sin(x_test)*SCALE)} (={math.sin(x_test):.6f})')
print(f'Error:  {abs(result/SCALE - math.sin(x_test)):.2e}')
```
