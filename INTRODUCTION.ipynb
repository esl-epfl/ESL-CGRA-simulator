# Test a ready kernel

```python
import sys
sys.path.append("src/")
from cgra import *

# Set the kernel name
kernel_name = "examples/convolution"
version = "_v1"
# Input data
clear_memory(kernel_name, version)
add_memory_region(kernel_name, 36464, [1, 2, 3, 4, 5, 6, 7, 8, 9], version)
add_memory_region(kernel_name, 100, [9], version)
# Set load and write addrs
load_addrs = [36464, 36464, 36464, 100]
store_addrs = [0, 0, 0, 4]
# Run the example
run(kernel_name, version=version, pr=["ROUT","OPS", "ALL_LAT_INFO"], load_addrs=load_addrs, store_addrs=store_addrs)
```
